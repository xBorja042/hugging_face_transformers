```python
print(" Q1 ")

from transformers import pipeline
classifier = pipeline('zero-shot-classification', model='roberta-large-mnli')

sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
print(sequence_to_classify, classifier(sequence_to_classify, candidate_labels))


sequence_to_classify = "the strong side of the team is the speed that makes them be always where they need to be to hit the point and win every competition"
candidate_labels = ['sports', 'cooking', 'weather']
print(sequence_to_classify, classifier(sequence_to_classify, candidate_labels))




sequence_to_classify = "The CEO had a strong handshake."
candidate_labels = ['male', 'female']
hypothesis_template = "This text speaks about a {} profession."
classifier(sequence_to_classify, candidate_labels, hypothesis_template=hypothesis_template)
print(sequence_to_classify, classifier(sequence_to_classify, candidate_labels))
```

```python
print("Q2")

from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
print(ner("My name is Sylvain and I work at Hugging Face in Brooklyn."))

print(ner("Magic The Gathering a board game developed by Richar Garfield in Philadelphia while he was studying at the Pensilvania University"))
```
