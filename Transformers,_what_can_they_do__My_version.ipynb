```python
input_quote = "For me there is no discussion betweeen Lebron and MJ. MJ is by far the best of the history due to the relevance of the moments in which he made his great performances despite the fact that "

def get_generated_text(input_quote:str, num_return_sequences: int=2, model: str="distilgpt2") -> list:
    print("Input quote is: \n ", input_quote)
    text_start = len(input_quote)
    generator = pipeline("text-generation", model=model)  
    generated_text_start = len(input_quote)
    text = generator(
        input_quote,
        max_length=60,
        num_return_sequences=num_return_sequences,
    )
    gen_text = [text[index]["generated_text"][text_start:] for index in range(num_return_sequences)]
    print(f"Generated text ---> {gen_text}")
    # return gen_text

get_generated_text(input_quote)   
```

```python
get_generated_text(input_quote, 2, "gpt2")   
```

```python
second_quote = "the spanish empire conquered america at the cost of"

print("DISTILGPT2: \n")
print(get_generated_text(second_quote))
print("GPT2: \n", get_generated_text(second_quote, 2, "gpt2"))
```
